````python
import pandas as pd
import sidetable as stb
````

````python
# Carregar dataset do Titanic
train = pd.read_csv("Titanic/train.csv")
train.head(3)
````

````python
train.info()
````

# FREQ

````python
train.stb.freq(['Pclass'])
````

````python
train.stb.freq(['Pclass'], thresh=60)
````

````python
train.stb.freq(['Pclass'], thresh=60, other_label = 'Outros')
````

````python
train.stb.freq(['Pclass'], value='Fare')
````

````python
train.stb.freq(['Survived'], value='Fare')
````

````python
train.stb.freq(['Pclass','Survived', 'Sex'])
````

````python
train.stb.freq(['Pclass','Embarked', 'Sex', 'SibSp'], 
              thresh=50, 
              other_label='Restante dos Grupos') 
````

````python
train.stb.freq(['Pclass','Embarked', 'Sex', 'SibSp'], 
              thresh=50, 
              other_label='Restante dos Grupos',
              style=True) 
````

````python
train.stb.freq(['Embarked'])
````

````python
train.shape
````

````python
train.fillna('NULL').stb.freq(['Embarked'])
````

# COUNTS

````python
train.stb.counts()
````

````python
train.stb.counts(exclude='object')
````

````python
train.stb.counts(include='object')
````

# MISSING

````python
train.stb.missing()
````

````python
train.stb.missing(style=True)
````

````python
train.stb.missing(style=True, clip_0=True)
````

# SUBTOTAL

````python
train[['Sex','Embarked', 'Age', 'Fare']].groupby(['Sex','Embarked']).mean()
````

````python
train[['Sex','Embarked', 'Age', 'Fare']].groupby(['Sex','Embarked']).mean().stb.subtotal()
````

# FLATTEN

````python
train.groupby(['Embarked', 'Sex']).agg({'Fare': ['median'], 'Age': ['mean']})
````

````python
train.groupby(['Embarked', 'Sex']).agg({'Fare': ['median'], 'Age': ['mean']}).stb.flatten()
````
