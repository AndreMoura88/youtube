```python
import numpy as np
import pandas as pd
  
data = {'Nome':['Andre', 'Bruno', 'Maria', 'Natalia' ,'Thiago', 'Vanessa'], 
        'Idade':[20, 24, np.nan, 19, np.nan, 20],
        'UF':['SP', 'MG', np.nan, 'MG', 'BA', np.nan]} 

df = pd.DataFrame(data) 
  
df 
```

```python
# é ou nao é missing?
df.isna()
```

```python
# contagem de missing por coluna
df.isna().sum()
```

```python
# contagem de nao-missing por coluna
df.info()
```

```python
df[pd.isna(df['UF'])] # traz todas as linhas em que UF é missing
```

```python
df[pd.isna(df['Idade'])] # traz todas as linhas em que Idade é missing
```

```python
df.fillna(0) # preenche os missing com 0
```

```python
df
```

Se eu quisesse preencher o missing com 0 em definitivo (alterar dataset original), teria que usar o parâmetro INPLACE. Desta forma:

df.fillna(0, inplace=True)

ou simplesmente atribuir o objeto fillna ao dataset original:

df = df.fillna(0)

```python
# preenchendo o missing com a media
df['Idade'].fillna(df['Idade'].mean())
```

```python
# preenchendo o missing com a mediana
df['Idade'].fillna(df['Idade'].median())
```

```python
# veja que continuamos sem alterar o dataset original, por enquanto
df
```

```python
valores = {'Idade': df['Idade'].mean(), 'UF': 'N/A'}
df.fillna(valores) 

# novamente, para alterar dataset original, use df.fillna(valores, inplace=True)
```

```python
df.dropna() # exclui todas as linhas que tem alguma coluna missing
```

```python
# novamente, se nao usar o inplace, o dataset original nao é alterado. Veja que df segue inalterado:
df
```

```python
df.dropna(subset=['Idade']) # exclui só as linhas que tem idade missing, nao importa outras colunas
```

```python
# novo dataset para ultimo exemplo:
 
data = {'Nome':['Andre', 'Bruno', np.nan, 'Natalia' ,'Thiago', 'Vanessa'], 
        'Idade':[20, 24, np.nan, 19, np.nan, 20],
        'UF':['SP', 'MG', np.nan, 'MG', 'BA', np.nan]} 

df = pd.DataFrame(data) 
  
df 
```
